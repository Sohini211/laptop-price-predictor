# WEB SCRAPING DATA FROM FLIPKART
### Web scraping is a technique to fetch data from websites. While surfing on the web, many websites donâ€™t allow the user to save data for personal use. One way is to manually copy-paste the data, which both tedious and time-consuming. Web Scraping is the automation of the data extraction process from websites. We will perform web scraping using the requests library and beautifulsoup library in Python. 

```python
import numpy as np
import pandas as pd
```

```python
#importing important libraries for web scraping
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

We can see that all the content of the page is under the div with class entry-content. We will use the find class. This class will find the given tag with the given attribute. In our case, it will find all the div having class as entry-content. 

```python
#we need the data of the containers written under<div>
#class name of each container is same
soupdata.findAll('div',{'class':'_2kHMtA'})
```

containers are separated by commas, here are the all 24 containers of the 1st page.

```python
containers=soupdata.findAll('div',{'class':'_2kHMtA'})
print(type(containers),len(containers))
```

```python
#checking dir of resultset
dir(containers)
```

```python
#since resulset is almost like list we use indexing 
#data of first container
containers[0]
```

we will fetch the following data
1) product name
2) current price
3) mrp
4) stars
5) emi
6) ratings

```python
f= open('laptops_info.csv (4)','wb')#file will open in excel
f.write('ProductName,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,Storage,ScreenSize,ImageURL\n'.encode())

#fetching product names
import re
for container in containers:
    
    product=container.findAll('div',{'class':'_4rR01T'})#here we can also use find as there is only one tag of this classname(find only fetch first tag with that class name)
    ProductName=product[0].text.split('-')[0].strip()#product[0] is a beautiful soup element-tag 
 #we just need the text before the first '-'   , the above data is string ,we can use split,we use strip to remove the space at the end
    
    
#fetching ratings    
    stars=container.find('div',{'class': '_3LWZlK'})
    try:
        Stars=stars.text
    except:
        Stars=0
     #finding ratings & reviews
    Rating=container.find('span',{'class':'_2_R_DZ'})
    try:
        ratRev=re.findall('\d+,?\d*',Rating.text)
        Ratings=ratRev[0].replace(',','')
        Reviews=ratRev[1].replace(',','')
        
    except:
        Ratings=0
        Reviews=0
    
    #finding current price 
    CurrentPrice=container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','').replace('â‚¹','')
    
    #finding MRP
    mrp=container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP=mrp.text.replace(',','').replace('â‚¹','')
    except:
        MRP=0
    #finding information about this product    
    info = container.findAll('li',{'class':'rgWa7D'})   
    Processor=info[0].text
    RAM=info[1].text
    try:
        Storage=info[3].text
    except:
        Storage=0
        
    
    ScreenSize=info[4].text
    
    Image=container.img
    ImageURL=Image.get('src')
    print(ProductName,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,Storage,ScreenSize,ImageURL)
    f.write(f"{ProductName},{Stars},{Ratings},{Reviews},{CurrentPrice},{MRP},{Processor},{RAM},{Storage},{ScreenSize},{ImageURL}\n".encode())
    
    print('\n')
f.close()  
```

### Preprocessing Data

```python
#fetching the brand name from product name column
output=[]
for i in range(0,971):
    brand = df['ProductName'][i].split()[0]
    output.append(brand)
                
```

```python
df=df.assign(Brandname=output)
```

```python
#fetching only numerical term from column "RAM"
output1=[]
for i in range(0,971):
    ram=str(df['RAM'][i]).split()[0]
    output1.append(ram)
  
```

```python
df=df.assign(RAM=output1)
```

```python
#fetching only numerical term from column "Storage"
output2=[]
for i in range(0,971):
    SSD=str(df['Storage'][i]).split()[0]
    output2.append(SSD)
 
```

```python
df=df.assign(Storage=output2)
```

```python
df['Processor'].value_counts()
```

```python
df['Cpu Name']=df['Processor'].apply(lambda x:" ".join(x.split()[0:3]))#grabbing 1st 3 words and make it string
```

```python
df.drop(columns='Processor',inplace= True)
```

```python
df.head()
```

```python
#fetching screensize information in cm
output3=[]
for i in range(0,971):
    Screensize=str(df['ScreenSize'][i]).split()[0]
    output3.append(Screensize)  
```

```python
df=df.assign(ScreenSize=output3)
```

```python
df['Cpu Name'].value_counts()
```

```python
# using fetch_processor function we would classify Cpu in major 9 classes
def fetch_processor(text):
    if text=='Intel Core i7'or text=='Intel Core i5'or text=='Intel Core i3' or text=='AMD Ryzen 3' or text=='AMD Ryzen 9'or text=='AMD Ryzen 5'or text=='AMD Ryzen 7'or text=='Intel Core i9' :
        return text
    else:
        return 'others'
```

```python
df['Cpu Brand']=df['Cpu Name'].apply(fetch_processor)
```

```python
df.drop(columns=['Cpu Name'],inplace=True)
```

```python
df['Cpu Brand'].value_counts()
```

```python
def fetch_processor2(text):
    if text=='AMD Ryzen 3':
        return 1
    elif text=='Intel Core i3':
        return 2
    elif text=='Intel Core i5' or text =='AMD Ryzen 5':
        return 3
    elif text=='Intel Core i7' or text =='AMD Ryzen 7':
        return 4
    elif text =='Intel Core i9':
        return 5
    elif text =='AMD Ryzen 9':
        return 6
    else:
        return 0
```

```python
df['Processor']=df['Cpu Brand'].apply(fetch_processor2)
```

```python
df.head()
```

```python
df['RAM'].value_counts()
```

```python
#using fetch_ram function we would classify RAM in major 5 classes
def fetch_ram(num):
    if num =='8' or num =='16' or num=='32' or num=='4':
        return num
    elif num == '16GB':
        return 16
    else:
        return 0
```

```python
df['RAM']=df['RAM'].apply(fetch_ram)
```

```python
df['Storage'].value_counts()
```

```python
#using fetch_storage function we would classify Storage in major 5 classes
def fetch_storage(num):
    if num =='512' or num =='256' or num=='128' or num=='64':
        return num
    elif num == '512GB':
        return 512
    else:
        return 0
```

```python
df['Storage']=df['Storage'].apply(fetch_storage)
```

```python
df['Stars'].value_counts()
```

```python
def fetch_stars(num):
    if num ==' Intel EVO Core i7 13th Gen':
        return 0
    else:
        return num
```

```python
df['Stars']=df['Stars'].apply(fetch_stars)
```
